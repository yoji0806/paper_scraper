## Functions

```python
def soup(url, parser="html.parser"):
    """Get a web page and parse
    
    Args:
        url (string): url of a web page.
        parser (string): parser available in bs4.

    Returns:
        soup_obj (bs4.BeautifulSoup): soup oject
    """
    
    response = requests.get(url)
    #to avoid character corruption
    response.encoding = response.apparent_encoding
    soup_obj = BeautifulSoup(response.text, parser)
    
    return soup_obj
```

## make Kotlin Code which convert JEF code to classification

```python
url = 'https://www.aeaweb.org/econlit/jelCodes.php?view=econlit'
soup_obj = soup(url)
```

```python
slice_start = soup_obj.text.find("A00\nGeneral Economics and Teaching")
slice_end = soup_obj.text.find("\n\t\t\t    \n\n\n\n\n\nEconLitSubscriptionsJournals")

print(f'slice soupobject.text from {slice_start} to {slice_end}')


txt_JEL = soup_obj.text[slice_start:slice_end].split(pattern1)
```

```python
'''
between code&classification lines there were some characters in the three patterns below, 
so split into one line each as an item of List, using the patterns.

- "\n\t\t\t    \n\n\n"
- "\n\t\t\t    \n\n\xa0\n\n"
- "\n\n\n\n"
'''

code_class_list = []

pattern1 = '\n\n\n\n'
pattern2 = '\n\t\t\t    \n\n\xa0\n\n'
pattern3 = '\n\t\t\t    \n\n\n'



for line_a in txt_JEL:
    line_b = line_a.split(pattern2)
    
    for line_c in line_b:
        line_d = line_c.split(pattern3)
        
        for code_class in line_d:
            code_class_list.append(code_class)
```

```python
for one_line in code_class_list:
    
    code, classification = one_line.split('\n')
    
    
    # print Kotlin code(snippet)
    print(f'"{code}" -> "{classification}"')
```
